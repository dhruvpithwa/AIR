```python
import geopandas as gpd
import json
```

```python
country=gpd.read_file('Indian_States.geojson')
```

```python
li=list(country.stname.unique())
```

```python
codes=["JK","HP","PB","UK","HR","CH","UP","RJ","AR","DL","AS","SK","NL","BR","WB","MP","ML","MN","JH","GJ","MZ","TR","CG","OD","MH","DD","DN",
       "AR","KA","PY","GA","TN","AN","KL","LD"]
```

```python
for i in range(0,len(li)):
  df=country[country.stname==li[i]]
  df=json.loads(df.to_json())
  s=codes[i]+".json"

  with open(s, 'w') as json_file:
    json.dump(df, json_file)
```
